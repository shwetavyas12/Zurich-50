```python
# Import necessary libraries
import pandas as pd

# Load users.csv file
from google.colab import files
uploaded = files.upload()  # This will prompt you to upload the file manually

# Read the CSV file into a DataFrame
users_df = pd.read_csv('users.csv')
```

```python
#Ques1-- Who are the top 5 users in Zurich with the highest number of followers? List their login in order, comma-separated.
import pandas as pd

# Sort the DataFrame by 'followers' in descending order and get the top 5
top_5_users = users_df.sort_values(by='followers', ascending=False).head(5)

# Extract the 'login' column values as a comma-separated string
top_5_logins = ",".join(top_5_users['login'].values)

print(top_5_logins)
```

```python
#Ques2-- Who are the 5 earliest registered GitHub users in Zurich? List their login in ascending order of created_at, comma-separated.

# Sort users by created_at in ascending order
earliest_users = users_df.sort_values(by='created_at', ascending=True).head(5)

# Get the login names in a comma-separated format
earliest_logins = ','.join(earliest_users['login'])
print("Earliest 5 users by registration:", earliest_logins)
```

```python
#Ques3-- What are the 3 most popular license among these users? Ignore missing licenses. List the license_name in order, comma-separated.

# Load repositories.csv
repositories_df = pd.read_csv('/content/repositories.csv')  # Update with the actual path if needed

# Filter out repositories with missing licenses
licenses_df = repositories_df[repositories_df['license_name'] != '']

# Count the occurrences of each license and get the top 3
top_licenses = licenses_df['license_name'].value_counts().head(3).index.tolist()

# Format the result as a comma-separated string
top_licenses_str = ",".join(top_licenses)
print("Top 3 most popular licenses:", top_licenses_str)
```

```python
# Ques4-- Which company do the majority of these developers work at?

# Count the occurrences of each company
company_counts = users_df['company'].value_counts()

# Get the company with the most occurrences
most_common_company = company_counts.index[0]

print(f"The company with the most developers is: {most_common_company}")
```

```python
#Ques5: Which programming language is most popular among these users?

# Count the occurrences of each language
language_counts = repositories_df['language'].value_counts()

# Get the language with the most occurrences
most_common_language = language_counts.index[0]

print(f"The most popular programming language is: {most_common_language}")
```

```python
# ques6: Which programming language is the second most popular among users who joined after 2020?

import pandas as pd

# Convert 'created_at' to datetime objects
repositories_df['created_at'] = pd.to_datetime(repositories_df['created_at'])

# Filter users who joined after 2020
recent_users = repositories_df[repositories_df['created_at'].dt.year > 2020]

# Count the occurrences of each language among recent users
language_counts = recent_users['language'].value_counts()

# Get the second most popular language
if len(language_counts) >= 2:
    second_most_popular_language = language_counts.index[1]
    print(f"The second most popular programming language among users who joined after 2020 is: {second_most_popular_language}")
else:
    print("Not enough data to determine the second most popular language.")
```

```python
#Ques7: Which language has the highest average number of stars per repository? use stargazers_count column in repositories_df

# Calculate the average number of stars per repository for each language
language_avg_stars = repositories_df.groupby('language')['stargazers_count'].mean()

# Find the language with the highest average stars
highest_avg_stars_language = language_avg_stars.idxmax()

print(f"The language with the highest average number of stars per repository is: {highest_avg_stars_language}")
```

```python
#Ques8: Let's define leader_strength as followers / (1 + following). Who are the top 5 in terms of leader_strength? List their login in order, comma-separated. use users_df

# Calculate leader_strength
users_df['leader_strength'] = users_df['followers'] / (1 + users_df['following'])

# Sort by leader_strength in descending order and get the top 5
top_5_leaders = users_df.sort_values(by='leader_strength', ascending=False).head(5)

# Extract logins and join them with commas
top_5_leader_logins = ','.join(top_5_leaders['login'])

# Print the result
top_5_leader_logins
```

```python
#Ques 9: What is the correlation between the number of followers and the number of public repositories among users in Zurich?
# Correlation between followers and repos (to 3 decimal places, e.g. 0.123 or -0.123), use users_df

import pandas as pd

# Filter users from Zurich
zurich_users = users_df[users_df['location'] == 'Zurich']

# Calculate the correlation
correlation = zurich_users['followers'].corr(zurich_users['public_repos'])

# Print the correlation rounded to 3 decimal places
print(f"{correlation:.3f}")
```

```python
#Ques 10: Does creating more repos help users get more followers? Using regression, estimate how many additional followers a user gets per additional public repository.
# Regression slope of followers on repos (to 3 decimal places, e.g. 0.123 or -0.123)

import pandas as pd
from scipy.stats import linregress

# Run linear regression on followers vs. public_repos
slope, intercept, r_value, p_value, std_err = linregress(users_df['public_repos'], users_df['followers'])

# Output the regression slope to 3 decimal places
print(f"Regression slope of followers on repos: {slope:.3f}")
```

```python
#Ques11:  Do people typically enable projects and wikis together? What is the correlation between a repo having projects enabled and having wiki enabled?
# Correlation between projects and wiki enabled (to 3 decimal places, e.g. 0.123 or -0.123)

import pandas as pd

# Load the data
repositories_df = pd.read_csv('repositories.csv')


# Calculate the correlation directly
correlation = repositories_df['has_projects'].astype(int).corr(repositories_df['has_wiki'].astype(int))

print(f"The correlation between having projects enabled and having a wiki enabled is: {correlation:.3f}")
```

```python
#Ques12: Do hireable users follow more people than those who are not hireable?
# Average of following per user for hireable=true minus the average following for the rest (to 3 decimal places, e.g. 12.345 or -12.345)

import pandas as pd

# Load the users data from the CSV file
users_df = pd.read_csv('users.csv')

# Filter hireable and non-hireable users
hireable_users = users_df[users_df['hireable'] == True]
non_hireable_users = users_df[users_df['hireable'].isna() | (users_df['hireable'] == False)]

# Calculate average following for both groups
average_hireable_following = hireable_users['following'].mean()
average_non_hireable_following = non_hireable_users['following'].mean()

# Calculate the difference
difference = average_hireable_following - average_non_hireable_following

# Print the result rounded to three decimal places
print(f'Difference in average following (hireable - non-hireable): {difference:.3f}')
```

```python
#Ques13: Some developers write long bios. Does that help them get more followers? What's the impact of the length of their bio (in Unicode words, split by whitespace) with followers? (Ignore people without bios)
# Regression slope of followers on bio word count (to 3 decimal places, e.g. 12.345 or -12.345)

users_with_bios = users_df.dropna(subset=['bio'])

# Calculate bio word counts
users_with_bios['bio_word_count'] = users_with_bios['bio'].astype(str).apply(lambda x: len(x.split()))

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(users_with_bios['bio_word_count'], users_with_bios['followers'])

# Print the slope rounded to 3 decimal places
print(f"{slope:.3f}")
```

```python
#Ques14: Who created the most repositories on weekends (UTC)? List the top 5 users' login in order, comma-separated
# Users login. use users_df

# Convert 'created_at' column to datetime objects
users_df['created_at'] = pd.to_datetime(users_df['created_at'])

# Extract the day of the week (0 = Monday, 6 = Sunday)
users_df['day_of_week'] = users_df['created_at'].dt.dayofweek

# Filter for weekend creations (Saturday and Sunday)
weekend_creations = users_df[(users_df['day_of_week'] == 5) | (users_df['day_of_week'] == 6)]

# Count repositories created by each user on weekends
weekend_repo_counts = weekend_creations['login'].value_counts()

# Get the top 5 users
top_5_weekend_creators = weekend_repo_counts.head(5).index.tolist()

# Convert the list to a comma-separated string
result = ','.join(top_5_weekend_creators)

# Print the result
result
```

```python
#Ques15: Do people who are hireable share their email addresses more often?
# [fraction of users with email when hireable=true] minus [fraction of users with email for the rest] (to 3 decimal places, e.g. 0.123 or -0.123). use users_df. approach in a different manner.

# Ensure 'hireable' is converted to boolean (True for hireable, False otherwise)
users_df['hireable'] = users_df['hireable'].fillna(False).astype(bool)

# Calculate the fraction of hireable users with an email
hireable_email_fraction = users_df[users_df['hireable']]['email'].notna().sum() / users_df[users_df['hireable']].shape[0]

# Calculate the fraction of non-hireable users with an email
non_hireable_email_fraction = users_df[~users_df['hireable']]['email'].notna().sum() / users_df[~users_df['hireable']].shape[0]

# Find the difference in fractions
email_difference = hireable_email_fraction - non_hireable_email_fraction
print(f"Difference in email sharing fraction: {email_difference:.3f}")
```

```python
#Ques16: Let's assume that the last word in a user's name is their surname (ignore missing names, trim and split by whitespace.) What's the most common surname? (If there's a tie, list them all, comma-separated, alphabetically)
# Most common surname(s). try again

import pandas as pd

# Assuming users_df is already loaded from the previous code

#Clean names, handle missing values
users_df['name'] = users_df['name'].astype(str).str.strip()
users_df = users_df[users_df['name'] != 'nan']


def get_surname(name):
    parts = name.split()
    if parts:
        return parts[-1]
    else:
        return None

users_df['surname'] = users_df['name'].apply(get_surname)

# Count surname occurrences and find most common
surname_counts = users_df['surname'].value_counts()

#Find max count
max_count = surname_counts.max()
most_common_surnames = surname_counts[surname_counts == max_count].index.tolist()
most_common_surnames.sort()

print(",".join(most_common_surnames))
```
